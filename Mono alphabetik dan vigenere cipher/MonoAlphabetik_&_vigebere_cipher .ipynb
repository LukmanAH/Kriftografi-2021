```python
# ------------------------#
#prosedur Mencetak Hasil
#-----------------------#
def cetak(a,b,c,pilihan):
    output = open('output.txt', 'w')
    print('')
    if pilihan == '1':
        print('P :' + a) 
        print('K :' + b) 
        print('C :' + c)
        output.write('Hasil Enkripsi: '+c)
    else:
        print('C :' + a) 
        print('K :' + b) 
        print('P :' + c)
        output.write('Hasil Enkripsi: '+c)
        
    output.close()

        
#-------------------------------#
#Prosedur Vigenere Cipher
#------------------------------#
def vigenere(text):
   
    index=0
    key2=''
    hasil=''
    
    print('')
    print('pilihan:')
    print('1. Enkripsi ')
    print('2. Dekripsi ')
    pilihan=input('Inputkan Pilihan[1 atau 2]: ')
    
    if pilihan != '1' and pilihan !='2':
        print("Invalid Input ,, Your Choices only  '1' or '2'")
        return ''
    
    key=input('Input Key (berupa kata) : ')
    
    for idk in range(len(text)):
        if(text[idk]==' '):
            key2+=' '
        else:
            key2+=key[index]
            index+=1
            if(index>len(key)-1):
                index-=len(key)
                
    #saya senang kriftografi = 24
    #pend ekpend ekpendekpen = 24
    
    
    
    for idk in range(len(text)):
        if text[idk] != ' ':
            if pilihan == '1':
                x=(ord(text[idk])-97+ord(key2[idk])-97)%26
                
                # x = 21
                
            else:
                x=(ord(text[idk])-ord(key2[idk]))%26
            x+=97
            hasil+=chr(x)
           #chr(119) = 'w'
            #'w'
        else:
            hasil+=' '
    
    cetak(text,key2,hasil,pilihan)

    
#----------------------------------#
#Prosedur Monoalfabetik Cipher Urut
#----------------------------------#
def monoalfabetik1(text):
    
    hasil=''
    
    print('')
    print('pilihan:')
    print('1. Enkripsi ')
    print('2. Dekripsi ')
    pilihan=input('Inputkan Pilihan[1 atau 2]: ')
    
    if pilihan != '1' and pilihan !='2':
        print("Invalid Input ,, Your Choices only  '1' or '2'")
        return ''
    
    key=int(input('Input Key (Berupa Angka): '))
    
    
    for idk in range(len(text)):
        if text[idk]!=' ':
            if(pilihan=='1'):  #enkripsi
                x=ord(text[idk])+key
                #x=(121)+3=124
                while (x > 122):
                    x-=26
            else:
                x=ord(text[idk])-key
                while (x < 97):
                    x+=26
                
            hasil+=chr(x)
            
        else:
            hasil+=' '
    
    cetak(text,str(key),hasil,pilihan)


    
#-------------------------------------#
#Prosedur Monoalfabetik Cipher Acak
#------------------------------------#
def monoalfabetik2(text):
    alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    hasil=''
    
    
    
    print('')
    print('pilihan:')
    print('1. Enkripsi ')
    print('2. Dekripsi ')
    pilihan=input('Inputkan Pilihan[1 atau 2]: ')
    
    if pilihan != '1' and pilihan !='2':
        print("Invalid Input ,, Your Choices only  '1' or '2'")
        return ''
    
    key=input('Masukin Key Berupa Kalimat:') 
    key2=''
    
    
    for idk in range(len(key)):
        if key[idk] !=' ':
            for idk2 in range(len(alphabet)):
                if key[idk]==alphabet[idk2]:
                    del alphabet[idk2]
                    key2+=key[idk]
                    break

    for idx in range(len(alphabet)): 
        key2+=alphabet[idx]
    
    alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    
    for idk in range(len(text)):
        if text[idk]!=' ':
            for idk2 in range(26):
                if pilihan == '1':
                    if text[idk]==alphabet[idk2] :    
                        hasil+=key2[idk2]
                else:
                    if text[idk]==key2[idk2] :    
                        hasil+=alphabet[idk2]                  
        else:
            hasil+=''
    
    cetak(text,key2,hasil,1)

#--------------#  
# main Program #
#--------------#
inputan = open('inputan.txt', 'r')

print('Pilih Algoritma: ')
print(' 1. Monoalfabetik Cipher [Urut]')
print(' 2. Monoalfabetik Cipher [Acak]')
print(' 3. Vigenere Cipher')

pilih=input('Masukan Pilihan[1, 2 atau 3] :')

if pilih == '1':
     monoalfabetik1(inputan.read())
elif pilih == '2':
    monoalfabetik2(inputan.read())
elif pilih == '3':
    vigenere(inputan.read())
else:
    print("Invalid Input ,, Your Choices only  '1' , '2' or '3'")
    
```
